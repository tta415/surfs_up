```python
import pandas as pd
```

## Hawaii_measurements.csv data wrangling

```python
# Import csv as dataframe and inspect
measurements = pd.read_csv("hawaii_measurements.csv", parse_dates = ["date"])
```

```python
measurements.head()
```

```python
# Decide what to do with Null Values -- decided to fill with 0; intepreted as 0 rainfall
measurements.fillna(0, inplace=True)
measurements['station'] = measurements['station'].astype('category')
```

```python
measurements.head()
```

```python
measurements.info()
```

```python
measurements.pivot(index = "date", columns = "station", values="prcp")
```

```python
# Export to "clean_" csv
measurements.to_csv("clean_hawaii_measuremnts.csv", index = False, encoding = "utf-8")
```
